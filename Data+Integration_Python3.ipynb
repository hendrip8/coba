```python
### Data integration
import pandas as pd
import numpy as np
```

```python
## Merging
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})

result1 = pd.merge(left, right, on='key')
```

```python
result1
```

```python
# Ordering variable/column
result1 = result1[['key', 'A', 'B', 'C','D']]
```

```python
result1
```

```python
# Multiple key
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                'key2': ['K0', 'K1', 'K0', 'K1'],
                'A': ['A0', 'A1', 'A2', 'A3'],
                'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                    'key2': ['K0', 'K0', 'K0', 'K0'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']})

result2 = pd.merge(left, right, on=['key1', 'key2'])
```

```python
result2
```

```python
# Merge methods (default how="inner")
result = pd.merge(left, right, how='left', on=['key1', 'key2'])
result
```

```python
result = pd.merge(left, right, how='right', on=['key1', 'key2'])
result
```

```python
result = pd.merge(left, right, how='outer', on=['key1', 'key2'])
result
```

```python
result = pd.merge(left, right, how='inner', on=['key1', 'key2'])
result
```

```python
## Joining
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                    index=['K0', 'K1', 'K2'])
right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                    index=['K0', 'K2', 'K3'])
result = left.join(right)
result
```

```python
result = left.join(right, how='outer')
result
```

```python
result = left.join(right, how='inner')
result
```

```python
#The data alignment here is on the indexes (row labels). 
#This same behavior can be achieved using merge plus additional arguments instructing it to use the indexes
result = pd.merge(left, right, left_index=True, right_index=True, how='outer')
result
```

```python
result = pd.merge(left, right, left_index=True, right_index=True, how='inner')
result
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                   'B': ['B0', 'B1', 'B2', 'B3'],
                   'key': ['K0', 'K1', 'K0', 'K1']})
right = pd.DataFrame({'C': ['C0', 'C1'],
                       'D': ['D0', 'D1']},
                      index=['K0', 'K1'])
result = left.join(right, on='key')
result
```

```python
## Concatenating
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                'B': ['B0', 'B1', 'B2', 'B3'],
                'C': ['C0', 'C1', 'C2', 'C3'],
                'D': ['D0', 'D1', 'D2', 'D3']})
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                'B': ['B4', 'B5', 'B6', 'B7'],
                'C': ['C4', 'C5', 'C6', 'C7'],
                'D': ['D4', 'D5', 'D6', 'D7']})
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                'B': ['B8', 'B9', 'B10', 'B11'],
                'C': ['C8', 'C9', 'C10', 'C11'],
                'D': ['D8', 'D9', 'D10', 'D11']})

frames = [df1, df2, df3]
result = pd.concat(frames)
result = pd.concat([df1,df2,df3])
```

```python
result
```

```python
# Add category
result = pd.concat(frames, keys=['X', 'Y', 'Z'])
result
```

```python
# Default join=outer (join all rows)
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                    index=[2, 3, 6, 7])

result = pd.concat([df1, df4], axis=0) #axis=1 is for concated by y-axis; default axis=0
result
```

```python
# Join = inner (join only rows that exist in both datasets)
result = pd.concat([df1, df4], axis=1, join='inner')
result
```

```python
# Join the exact rows in df1
result = pd.concat([df1, df4], axis=1, join_axes=[df1.index])
result
```

```python
## Concatenating using append
result = df1.append(df2) #default axis=0
result
```

```python
result = df1.append(df4)
result
```

```python
result = df1.append([df2, df3]) #multiple datasets to concatenate
result
```

```python
# For DataFrames which dont have a meaningful index, 
#you may wish to append them and ignore the fact that they may have overlapping indexes.
result = pd.concat([df1, df4], ignore_index=True)
result
```

```python
result = df1.append(df4, ignore_index=True)
result
```

```python
# Appending rows to a DataFrame
s2 = pd.Series(['X0', 'X1', 'X2', 'X3'], index=['A', 'B', 'C', 'D'])
result = df1.append(s2, ignore_index=True)
result
```

```python
dicts = [{'A': 1, 'B': 2, 'C': 3, 'X': 4}, #dictionary
         {'A': 5, 'B': 6, 'C': 7, 'Y': 8}]
result = df1.append(dicts, ignore_index=True)
result
```
