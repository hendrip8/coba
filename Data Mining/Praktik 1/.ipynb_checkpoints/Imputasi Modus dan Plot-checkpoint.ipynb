# Mengatasi Missing

```python
#imputasi mean
movie['duration']=movie['duration'].fillna(movie['duration'].mean())
movie['facenumber_in_poster']=movie['facenumber_in_poster'].fillna(movie['facenumber_in_poster'].mean())
movie['gross']=movie['gross'].fillna(movie['gross'].mean())
movie['budget']=movie['budget'].fillna(movie['budget'].mean())
movie['num_user_for_reviews']=movie['num_user_for_reviews'].fillna(movie['num_user_for_reviews'].mean())
movie['num_critic_for_reviews']=movie['num_critic_for_reviews'].fillna(movie['num_critic_for_reviews'].mean())
movie['director_facebook_likes']=movie['director_facebook_likes'].fillna(movie['director_facebook_likes'].mean())
movie['actor_1_facebook_likes']=movie['actor_1_facebook_likes'].fillna(movie['actor_1_facebook_likes'].mean())
movie['actor_2_facebook_likes']=movie['actor_2_facebook_likes'].fillna(movie['actor_2_facebook_likes'].mean())
movie['actor_3_facebook_likes']=movie['actor_3_facebook_likes'].fillna(movie['actor_3_facebook_likes'].mean())
movie.head()
```

```python
#imputasi median
movie['duration']=movie['duration'].fillna(movie['duration'].median())
movie['facenumber_in_poster']=movie['facenumber_in_poster'].fillna(movie['facenumber_in_poster'].median())
movie['gross']=movie['gross'].fillna(movie['gross'].median())
movie['budget']=movie['budget'].fillna(movie['budget'].median())
movie['num_user_for_reviews']=movie['num_user_for_reviews'].fillna(movie['num_user_for_reviews'].median())
movie['num_critic_for_reviews']=movie['num_critic_for_reviews'].fillna(movie['num_critic_for_reviews'].median())
movie['director_facebook_likes']=movie['director_facebook_likes'].fillna(movie['director_facebook_likes'].median())
movie['actor_1_facebook_likes']=movie['actor_1_facebook_likes'].fillna(movie['actor_1_facebook_likes'].median())
movie['actor_2_facebook_likes']=movie['actor_2_facebook_likes'].fillna(movie['actor_2_facebook_likes'].median())
movie['actor_3_facebook_likes']=movie['actor_3_facebook_likes'].fillna(movie['actor_3_facebook_likes'].median())
movie.head()
```
