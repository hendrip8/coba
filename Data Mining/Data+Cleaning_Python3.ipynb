```python
    # Smoothing by regression
# Import the library
from sklearn import linear_model
#from sklearn.linear_model import LinearRegression

# Create linear regression object
regr = linear_model.LinearRegression()
x = movie['actor_1_facebook_likes']
y = movie['actor_2_facebook_likes']

# Regress
regr.fit(x.to_frame(), y.to_frame())

# Make predictions
movie['actor_2_facebook_likes_pred'] = regr.predict(x.to_frame())
```
